```python
#creating a scatter plot of the "Age" and "Employee Salary"

plt.figure(figsize=(11,6))

sns.scatterplot(data= data, x= 'age', y='salary', color='blue',
                edgecolor= 'linen', alpha=0.5)

plt.title("Scatterplot of Employee Age Vs Employee salary")
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')
plt.show
```

```python
#observation: 
#there are some outlier after age 50, infact age increase salary increase.
```

```python
#Creating scatter plot of the Healthy_eating and active_lifestyle

plt.figure(figsize=(11,6))

sns.scatterplot(data= data, x= 'healthy_eating', y= 'active_lifestyle', color= 'red',
                edgecolor= 'linen', alpha=0.5)

plt.title("Scatterplot of Healthy eating Vs Active lifestyle")
plt.xlabel('Healthy Eating')
plt.ylabel('Acitve Lifestyle')
plt.show()
```

```python
#Observation"
#Both are directly proportional to each other i.e healthy eating increase as per active lifestyle 
```

```python
#Create a scatter plot of the healthy_eating and salary 
plt.figure(figsize=(11,6))

sns.scatterplot(data= data, x= 'healthy_eating', y= 'salary', color= 'k',
                edgecolor='linen', alpha=0.5)

plt.title("Scatter plot Healthy eating Vs Salary")
plt.xlabel('Healthy eating')
plt.ylabel('Salary')
plt.show()
```

```python
#Observation
#slope is form and both healthy eating directly proportional to salary.
```

```python
# Create count plot or detail data of groups 

plt.figure(figsize= (11,6))

sns.countplot(data=data, x='groups', edgecolor= 'linen', alpha=0.7)

plt.title("Count Plot of Groups")
plt.xlabel("Groups")
plt.ylabel("Count")
plt.show()
```

```python
#Observation: most employees are either Group A or Group,
#Group A having max frequency
```

```python
# Creating Histogram of the salary

plt.figure(figsize=(11,7))

sns.histplot(data= data, x= 'salary', color= 'orange',
             edgecolor='linen', alpha=0.5, bins=10)

plt.title("Histogram of Salary")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
#Observation: salary is not uniform, not even so no conclusion can be made
#there are employees with salaries on either extremities of the histogram
#max employees are between 1000-2000.
```
