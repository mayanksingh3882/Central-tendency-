```python
#Task1: The features in the data set are: Different acids and their quality
#Task2: View the distributions of the various features in the data set and calculate their central tendencies
```

```python
#Create a histogram of the "fixed acidity" feature

plt.figure(figsize = (11,6))

sns.histplot(data = data,x= 'fixed acidity', color='orange',edgecolor= 'linen', alpha=0.5, bins=5)

plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel('Count')
plt.show
```

```python
#Observation -> the histogram is normally distributed
#the max. count of values for fixed acidity lies in between 6 to 8.
```

```python
#Let's see the measure of central tendency through mean, median and mode.
```

```python
data['fixed acidity'].mean()  #mean
```

```python
round(data['fixed acidity'].mean(),2)
```

```python
data['fixed acidity'].median() #median
```

```python
#showcasing the mean median in the above histogram
plt.figure(figsize =(11,6))

sns.histplot(data = data, x='fixed acidity', color= 'orange',
             edgecolor = 'linen', alpha=0.5, bins=5)

plt.title("Histogram of fixed acidity")
plt.xlabel('Fixed acidity')
plt.ylabel('Count')

plt.vlines(data['fixed acidity'].mean(), ymin=0, ymax=4000, colors='blue', label='Mean')
plt.vlines(data['fixed acidity'].median(), ymin= 0, ymax= 4000, colors='red', label='Median')
plt.legend()
plt.show
```

```python
#Observations
#We can see that mean and median are clear representative of the data
#Mean and median are very close to each other.
#We can choose either of the parameters say mean as the measure of central tendency
```

```python
#Create a histogram of the Volatile acidity feature

plt.figure(figsize=(11,6))

sns.histplot(data= data, x='volatile acidity', color= 'green',
             edgecolor= 'linen', alpha=0.5, bins=5)
plt.title('Histogram of Volatile Acidity')
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')
plt.show
```

```python
#here the distribution is skewed a little towards the right
#As we have seen skewness, therefore we can check the distribution using distplot function.
```

```python
#Plotting distplot using 'Volatile acidity' feature

plt.figure(figsize = (11,6))

sns.distplot(data['volatile acidity'], color ='blue')

plt.title("Distplot of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')
plt.show()
```

```python
#Observation - the above plot shows the normal distribution

#The normal distribution is describe by the mean and the standard deviation
# Oftenly referred as bell curve because of the shape.
# median and mean are equal, it has only one mode, it is sysmmetric meaning it decreases the same
#amount on the left and the right of the centre.
```

```python
#Calcualte the skewness of  'volatile acidity'
data['volatile acidity'].skew()
```

```python
#  >0 i.e it is positively skew
```

```python
data['volatile acidity'].mean()   #mean
```

```python
data['volatile acidity'].median()   #median
```

```python
#showing mean median in the histplot
plt.figure(figsize= (11,6))

sns.histplot(data= data, x= 'volatile acidity', color= 'green',
             edgecolor= 'linen', alpha=0.5, bins= 5)

plt.title("Histogrema of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel('Density')
plt.vlines(data['volatile acidity'].mean(), ymin=0, ymax=4000, colors='blue', label='Mean')
plt.vlines(data['volatile acidity'].median(), ymin=0, ymax= 4000, colors= 'red' , label= 'Median')
plt.legend()
plt.show
```

```python
#Obseravation -> Mean and median are close to each other and the difference between them is very small
#we can choose the mean as the measure of the central tendency here.
```
